<h2 center>Part of Speech en Spacy</h2>

Cargar el corpus.json

```python
import json

with open('recursos/corpus_stopword_procesado.json', 'r', encoding='utf-8') as file:
    corpus = json.load(file)
```

Procesar cada documento del corpus utilizando Spacy:

```python
import spacy

# Cargar el modelo de Spacy
nlp = spacy.load("es_core_news_sm")
```

```python
# Selecci√≥n de un documento (0-35)
indice = 0

# Procesar el documento seleccionado
doc = nlp(corpus[indice]['contenido_procesado'])
```

```python
# imprimir los tokens del documento, sus etiquetas POS
for token in doc:
    print(token.text, "|", token.pos_, "|", spacy.explain(token.pos_))
```

Filtrar los tokens por tipo de palabra (SPACE, PUNCT, X):

```python
filtered_tokens = []
doc = nlp(corpus[indice]['contenido_procesado'])

for token in doc:
    if token.pos_ not in ['SPACE', 'PUNCT', 'X']:
        filtered_tokens.append(token.text)
```

```python
# Imprimir los tokens filtrados
from collections import Counter
word_freq = Counter(filtered_tokens)

for word, freq in word_freq.items():
    print(word, freq)
```
